```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
N = 100
D = 2
K = 3
X = np.zeros((N*K, D))
y = np.zeros(N*K, dtype='uint8')
```

```python
for j in range(K):
    ix = range(N*j,N*(j+1))
    r = np.linspace(0.0,1,N) # radius
    t = np.linspace(j*4,(j+1)*4,N) + np.random.randn(N)*0.2 # theta
    X[ix] = np.c_[r*np.sin(t), r*np.cos(t)]
    y[ix] = j

plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
W = 0.01 * np.random.randn(D, K)
b = np.zeros((1, K))
reg = 1e-3
learning_rate = 1e-0
```

```python
num_samples = X.shape[0]

for epoch in range(500):
    # calculating the scores and the loss
    scores = X.dot(W) + b
    exp_scores = np.exp(scores)
    prob = exp_scores / np.sum(exp_scores, axis=1, keepdims=True)
    correct_prob = -np.log(prob[np.arange(num_samples), y])
    loss = np.sum(correct_prob) / num_samples
    loss += 0.5 * reg * np.sum(W * W)
    
    if epoch % 50 == 0:
        print('The loss is {}'.format(loss))
    
    # calculating the gradients
    prob[np.arange(num_samples), y] -= 1
    prob /= num_samples
    dW = X.T.dot(prob)
    dW = dW + reg * W
    db = np.sum(prob, axis=0, keepdims=True)
        
    # updating the weight and regularization
    W += -learning_rate * dW
    b += -learning_rate * db
    
```

```python
scores = X.dot(W)
y_pred = np.argmax(scores, axis=1)
train_accuracry = np.mean(y == y_pred)
print('The training accuracy is {}'.format(train_accuracry))
```

```python
H = 100
W1 = 0.01 * np.random.randn(D, H)
b1 = np.zeros((1, H))
W2 = 0.01 * np.random.rand(H, K)
b2 = np.zeros((1, K))
```

```python
reg = 1e-3
learning_rate = 1e-0

for epoch in range(10000):
    # calculating the scores and loss
    s1 = np.maximum(0, X.dot(W1) + b1) # N x H
    s2 = s1.dot(W2) + b2 
    prob = np.exp(s2) / np.sum(np.exp(s2), axis=1, keepdims=True) # N x K
    correct_prob = prob[range(num_samples), y]
    loss = np.sum(-np.log(correct_prob)) / num_samples + 0.5 * reg * (np.sum(W1*W1) + np.sum(W2 * W2))
    
    if epoch % 100 == 0:
        print("Iteration: {}, loss = {}".format(epoch, loss))

    # calculating the gradients
    dprob = prob # N x K
    dprob[range(num_samples), y] -= 1
    dprob /= num_samples
    dW2 = s1.T.dot(dprob) + reg * W2 # H x K
    db2 = np.sum(dprob, axis=0, keepdims=True) # K x 1
    ds1 = dprob.dot(W2.T) # N x H
    ds1[s1 <= 0] = 0
    dW1 = X.T.dot(ds1) + reg * W1
    db1 = np.sum(ds1, axis=0, keepdims=True)

    # updating the weights
    W2 += -learning_rate * dW2
    b2 += -learning_rate * db2
    W1 += -learning_rate * dW1
    b1 += -learning_rate * db1
```

```python
scores = np.maximum(0, X.dot(W1) + b1).dot(W2) + b2
y_pred = np.argmax(scores, axis=1)
print("The training accuracy is {}".format(np.mean(y == y_pred)))
```
